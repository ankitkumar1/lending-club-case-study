```python
#Load the necessary Libraries
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns',100)

loan = pd.read_csv('Inferential-Statistics-Student-Loan.csv')
loan.head()
```

```python
loan['Recovery (%)'] = loan['Recovery (%)'].apply(lambda x: x.rstrip('%'))
loan
```

```python
loan = loan.drop(['Unnamed: 4', 'Unnamed: 5'], axis=1)
loan
```

```python
loan.isnull().sum()
```

```python
loan.dtypes
```

```python
loan['Recovery (%)'] = loan['Recovery (%)'].astype('float64')
loan.dtypes
```

```python
loan['recovery_amt'] = (loan['Exposure at Default (in lakh Rs.)']*loan['Recovery (%)'])/100
loan['balance_amt'] = loan['Exposure at Default (in lakh Rs.)'] - loan['recovery_amt']
loan['expected_loss'] = loan['balance_amt'] * loan['Probability of Default']
loan
```
